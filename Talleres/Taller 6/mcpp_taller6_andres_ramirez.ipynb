```python
#7.4.1. En este ejercicio no entendi muy bien si se tocaba mostrar los 365 días,
#por ello no se notan los valores en x, sin embargo, el punto siguiente (7.4.1),
#muestra lo que considero unifica los punto 7.4.1. y 7.4.2
import matplotlib.dates as mdates
import datetime as dt
base = dt.datetime(2020,3,1)
dates = np.array([base + dt.timedelta(days=i) for i in range(365)])
y = np.random.rand(len(dates))
plt.gca().xaxis.set_major_formatter(mdates.DateFormatter('%d/%m/%Y'))
plt.gca().xaxis.set_major_locator(mdates.DayLocator())
plt.plot(dates,y)
plt.gcf().autofmt_xdate()
```
